```python
import shutil

import pandas as pd

from huggingface_hub import hf_hub_download
from tqdm import tqdm
```

```python
train_labels[:10000]
```

```python
split = 'train_fixed'
batch_size = 5
start_at = 0

repo_id = 'ibrahimhamamci/CT-RATE'
directory_name = f'dataset/{split}/'
hf_token = 'token'

data = train_labels[:1000]

for i in tqdm(range(start_at, len(data), batch_size)):

    data_batched = data[i:i+batch_size]

    for name in data_batched['VolumeName']:
        folder1 = name.split('_')[0]
        folder2 = name.split('_')[1]
        folder = folder1 + '_' + folder2
        folder3 = name.split('_')[2]
        subfolder = folder + '_' + folder3
        subfolder = directory_name + folder + '/' + subfolder

        hf_hub_download(repo_id=repo_id,
            repo_type='dataset',
            token=hf_token,
            subfolder=subfolder,
            filename=name,
            cache_dir='./',
            local_dir='data_volumes',
            )
```

```python
train_datalist = []

for name in tqdm(train_labels[:1000].VolumeName):
    folder1 = name.split('_')[0]
    folder2 = name.split('_')[1]
    folder = folder1 + '_' + folder2
    folder3 = name.split('_')[2]
    subfolder = folder + '_' + folder3
        
    train_datalist.append({
        'name': name,
        'image': f'data_volumes/dataset/train_fixed/{folder}/{subfolder}/{name}',
        'mask': f'data_volumes/dataset/ts_seg/ts_total/train_fixed/{folder}/{subfolder}/{name}',
        'labels': train_labels[train_labels['VolumeName'] == name].to_numpy()[:, 1:][0].tolist(),
        'labels_human': train_labels.columns.to_list()[1:],
        'report': {
            'finding': train_reports[train_reports['VolumeName'] == name][['Findings_EN', 'Impressions_EN']].to_numpy()[0].tolist()[0],
            'impression': train_reports[train_reports['VolumeName'] == name][['Findings_EN', 'Impressions_EN']].to_numpy()[0].tolist()[1]
        },
        'metadata': train_metadata[train_metadata['VolumeName'] == name].iloc[0].to_dict()
        
    })
```

```python
valid_datalist = []

for name in tqdm(valid_labels[:50].VolumeName):
    folder1 = name.split('_')[0]
    folder2 = name.split('_')[1]
    folder = folder1 + '_' + folder2
    folder3 = name.split('_')[2]
    subfolder = folder + '_' + folder3
        
    valid_datalist.append({
        'name': name,
        'image': f'data_volumes/dataset/valid_fixed/{folder}/{subfolder}/{name}',
        'mask': f'data_volumes/dataset/ts_seg/ts_total/valid_fixed/{folder}/{subfolder}/{name}',
        'labels': valid_labels[valid_labels['VolumeName'] == name].to_numpy()[:, 1:][0].tolist(),
        'labels_human': valid_labels.columns.to_list()[1:],
        'report': {
            'finding': valid_reports[valid_reports['VolumeName'] == name][['Findings_EN', 'Impressions_EN']].to_numpy()[0].tolist()[0],
            'impression': valid_reports[valid_reports['VolumeName'] == name][['Findings_EN', 'Impressions_EN']].to_numpy()[0].tolist()[1]
        },
        'metadata': valid_metadata[valid_metadata['VolumeName'] == name].iloc[0].to_dict()
        
    })
```
